## Problem Statement
Build a decision tree classifier to predict whether a customer will purchase a product or service based on their demographic and behavioral data. Use a dataset such as the Bank Marketing dataset from the UCI Machine Learning Repository.

## Data Overview:

The dataset to be employed in this task is the Bank Marketing Dataset sourced from the UCI Machine Learning Repository. This dataset constitutes a 10% sample of the comprehensive Bank Marketing Dataset. Comprising 20 features and 1 label, this dataset pertains to direct marketing endeavors of a Portuguese banking institution. The outreach initiatives were predominantly conducted via phone calls, often necessitating multiple interactions with the same client to ascertain their subscription decision for a bank term deposit ('yes' or 'no').

## Features of the Dataset:

### Input variables:

##### Bank client data:
   Age (Numeric)
Job: Type of occupation (Categorical: 'admin.', 'blue-collar', 'entrepreneur', ... 'unemployed', 'unknown')
Marital: Marital status (Categorical: 'divorced', 'married', 'single', 'unknown'; 'divorced' encompasses divorced or widowed status)
Education: Educational background (Categorical: 'basic.4y', 'basic.6y', ... 'university.degree', 'unknown')
Default: Credit in default? (Categorical: 'no', 'yes', 'unknown')
Housing: Housing loan existence? (Categorical: 'no', 'yes', 'unknown')
Loan: Personal loan presence? (Categorical: 'no', 'yes', 'unknown')
##### Related with the last contact of the current campaign:
Contact: Communication method (Categorical: 'cellular', 'telephone')
Month: Last contact month (Categorical: 'jan', 'feb', ..., 'nov', 'dec')
Day of Week: Last contact day of the week (Categorical: 'mon', 'tue', ..., 'fri')
Duration: Duration of last contact in seconds (Numeric). It is noteworthy that this attribute significantly influences the outcome (e.g., duration=0 leads to y='no'). However, this duration is only ascertainable post-call, rendering it suitable for benchmarking purposes and potentially dispensable for constructing a realistic predictive model.
##### Other attributes:
Campaign: Count of contacts in this campaign for this client (Numeric, including last contact)
Pdays: Days elapsed since the client's last contact from a prior campaign (Numeric; 999 implies no prior contact)
Previous: Count of contacts executed before this campaign for this client (Numeric)
Poutcome: Result of the preceding marketing campaign (Categorical: 'failure', 'nonexistent', 'success')
##### Social and economic context attributes
Emp.var.rate: Quarterly employment variation rate (Numeric)
Cons.price.idx: Monthly consumer price index (Numeric)
Cons.conf.idx: Monthly consumer confidence index (Numeric)
Euribor3m: Daily euribor 3-month rate (Numeric)
Nr.employed: Quarterly employee count (Numeric)

##### Output variable (desired target):
y: Client subscription to a term deposit (Binary: 'yes', 'no')
The classification objective revolves around predicting whether the client will subscribe ('yes'/'no') to a term deposit (variable 'y').

### Plotting BoxPlot and Checking for Outliers

```python
df.plot(kind='box', subplots=True, layout=(2, 5), figsize=(20, 10), color='#1F618D') 
plt.show()
```


Through this Plot we can see there are 3 columns having outliers i.e.. 'Age', 'duration' and 'Campaign'. So, we will remove these outliers using Interquantile Range.

```python
# Removing outliers

column = df[['age','campaign','duration']]
q1 = np.percentile(column, 25)
q3 = np.percentile(column, 75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
df[['age','campaign','duration']] = column[(column > lower_bound) & (column < upper_bound)]
```

```python
# Plotting boxplot after removing outliers

df.plot(kind='box', subplots=True, layout=(2,5),figsize=(20,10),color='#808000')
plt.show()
```

### Label Encoding

```python
# Conversion of categorical columns into numerical columns using label encoder.
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
df_encoded = df1.apply(lb.fit_transform)
df_encoded
```

### Checking for target variable

```python
df_encoded['deposit'].value_counts()
```

### Selecting Independent and Dependent Variables

```python
x = df_encoded.drop('deposit',axis=1)  # independent variable
y = df_encoded['deposit']              # dependent variable
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

### Splitting the dataset into Train and Test datasets

```python
from sklearn.model_selection import train_test_split
```

```python
print(4119*0.25)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Creating a function to compute Confusion Matrix, Classification Report and to generate training and testing scores

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
def eval_model(y_test,y_pred):
    acc = accuracy_score(y_test,y_pred)
    print('Accuracy_Score',acc)
    cm = confusion_matrix(y_test,y_pred)
    print('Confusion Matrix\n',cm)
    print('Classification Report\n',classification_report(y_test,y_pred))

def mscore(model):
    train_score = model.score(x_train,y_train)
    test_score = model.score(x_test,y_test)
    print('Training Score',train_score)  # Training Accuracy
    print('Testing Score',test_score)    # Testing Accuracy
```

### Decision Tree Classifier

```python
# Decision Tree library
from sklearn.tree import DecisionTreeClassifier
```

```python
# Building Decision Tree Classifier Model
dt = DecisionTreeClassifier(criterion='gini',max_depth=5,min_samples_split=10)
dt.fit(x_train,y_train)
```

```python
# Evaluating training and testing accuracy
mscore(dt)
```

```python
# Generating prediction
ypred_dt = dt.predict(x_test)
print(ypred_dt)
```

```python
# # Evaluate the model - confusion matrix, classification Report, Accuaracy
eval_model(y_test,ypred_dt)
```
